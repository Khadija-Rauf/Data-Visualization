```python
df.Class.value_counts()
```

```python
# df["Class"].value_counts().plot(kind = "bar")
figure, ax = plt.subplots(figsize=(8,6))
sns.countplot(x="Class", data=df , ax=ax, palette = "bright")
plt.title('Class Occurences')
plt.show()
```

```python
figure.savefig(pwd + "/Charts/Class_BarChart.png")
```

```python
# To see students's absence days 
figure, ax = plt.subplots(figsize=(8,6))
sns.countplot(x="StudentAbsenceDays", data=df , hue = "Class", ax=ax, palette = "bright")
plt.title('Students Absence w.r.t Class')
plt.show()
```

```python
figure.savefig(pwd + "/Charts/Student_BarChart.png")
```

```python
d = pd.melt(df,id_vars='Class',value_vars=['VisITedResources','raisedhands','AnnouncementsView'])
```

```python
figure, ax = plt.subplots(figsize=(15,8))
sns.swarmplot(x='variable',y='value',hue='Class' , data=d, palette='bright')
sns.set(rc = {'figure.figsize':(15,8)})
plt.title('Students')
plt.show()
```

```python
# Saving plot in folder
figure.savefig(pwd + "/Charts/Students.png")
```

```python
figure, ax = plt.subplots(figsize=(8,6))
sns.countplot(x="Class", data=df, hue="gender" , ax=ax, palette = "bright")
plt.title('Gender Comparison')
plt.show()
```

```python
# Saving plot in folder
figure.savefig(pwd + "/Charts/Gender Comparison.png")
```
